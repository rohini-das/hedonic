```python
from itertools import chain, combinations
import random
import math
import numpy as np
from copy import deepcopy

def all_subsets(AA):
    
    def powerset(A):
        length = len(A)
        return {
        frozenset({e for e, b in zip(A, f'{i:{length}b}') if b == '1'})
        for i in range(2 ** length)}

    s = set()
    s = powerset(AA)
    l = list(s)
    l1 = [set(x) for x in l]
    l2 = [i for i in l1 if i != set()]
    #print(list(sub_sets))1,2,3,4
    return l2

s1 = [x for x in all_subsets({1,2,3,4})]
print(s1)


```

```python
def makegame(n):
    pref = dict((i, []) for i in n)
    util = dict((i, []) for i in n)
    for i in n:
        S_i = [x for x in all_subsets(n) if i in x]
        values = np.random.exponential(1, len(S_i))
        values = values.tolist()
        #print(S_i)
#         print(values)
#         print("\n")
        S_i.sort(key = len, reverse = True)
        #print(S_i)
        values.sort(reverse = True)
#         print(values)
#         print("\n")
        pref[i] = S_i
        util[i] = values
        
#     print(pref)
#     print("\n")
#     print(util)
    d = dict()
    for i in n:
        d[i] = [pref[i], util[i]]
#         print(d[i])
#         print("\n")
#     print(d)
        
        
    return d
```

```python
def shuffle(game, n):
    for i in n:
        l1 = (game[i])[0]
        l2 = (game[i])[1]
        random.shuffle(l1)
        vals = np.random.exponential(1, len(l2))
        vals = vals.tolist()
        vals.sort(reverse = True)
#         zipped = zip(l1, l2)
#         zipped = list(zipped)
#         zshuff = random.shuffle(zipped)
#         X_shuff, Y_shuff = zip(*zipped)
#         X_shuff = list(X_shuff)
#         Y_shuff = list(Y_shuff)
        (game[i])[0] = l1
        (game[i])[1] = vals
        
    return game
        
    
```
